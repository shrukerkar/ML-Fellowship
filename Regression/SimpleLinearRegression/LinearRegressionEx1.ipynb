```python
def cost_function(x_train,y_train,w,b):
    m=len(x_train)
    total_error=0
    for i in range(m):
        total_error+= (y_train[i]-(x_train[i]*w+b)**2)
        return total_error/m
```

```python
def new_weight(x_train,y_train,w,b,learning_rate):
    dw=0
    db=0
    m=len(x_train)
    for i in range(m):
        dw+= -2 * x_train[i]*(y_train[i]-(w*x_train[i]+b))
        db+=-2*(y_train[i]-(w*x_train[i]+b))
    w=(dw/m)*learning_rate
    b=(db/m)*learning_rate
    return w,b

                
```
